## What is EBM
First, we'll describe what EBM is.
EBM is an effort to make medical decisions accountable to the scientific literature.

One major challenge to this effort is that the standard approach to science relies on carefully constructed and controlled experiments.
Our patients are never carefully constructed nor carefully controlled.
So the relationships we study in experiments can often be quite different when the real world is taken into account.
We explore this idea in more detail over in a [separate notebook]().

Ultimately, the tools of EBM are designed around the idea that we're studying the interaction between two degrees of freedom.
In a system with n pieces, that's $n x (n-1)$ combinations of pairs.
Each of those combinations would need an experiment.
That's daunting.
But what's more daunting is that when you're performing an experiment, you have to keep all the other $(n-2)$ variables *constant*.
That's tough enough as it is, but before you *do* that you have to decide *what values* do you set each variable constant to?

If each variable has one of five options, then there are $5^{n-2}$ possibilities.
The relationship between your independent and dependent variables **may depend on what the other variables are fixed at**.
And that, right there, is the rub.

## What is ML
As we've described [elsewhere](), ML is the ability to study patterns across many degrees of freedom.

Machine learning (ML) is the mathematically rigorous way of finding patterns, whether they're simple or complex.
Simple patterns would be "Is A related to B?".
Complex patterns would be "How do A,B,C,D,E,F and G interact with each other?"

Much of Science, and all of EBM, is focused on identifying simple patterns.
ML, in contrast, assumes complex patterns and lets the data tell us if a simple pattern is enough to explain our data.
When talking about clinical problems and clinical patients, we're never dealing with simple patterns.
This is where ML promises to revolutionize the way we study physiology and improve therapies.

We'll use simple examples from medicine to help clarify why ML is such a powerful way to study physiology and pathophysiology.

```python
G = nx.Graph()
for ii in range(5):
    G.add_node(ii)
```

```python
for ii in range(5):
    for jj in range(5):
        if ii != jj:
            G.add_edge(ii,jj)
```
