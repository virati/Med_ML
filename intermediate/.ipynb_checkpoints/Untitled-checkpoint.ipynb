# What nonlinearities do to noise

Nonlinear relationships are complicated because they (a) amplify small bits of noise and (b) there are so many different types of nonlinearities.

In this notebook we'll explore the effect that nonlinearities have on noise

```python
import numpy as np
import matplotlib.pyplot as plt

#set up uniform noise on a 0-function
x = np.linspace(-10,10,1000)
y = 0 + np.random.uniform(-1.0,1.0,size=x.shape)
plt.scatter(x,y)
plt.hist(y)
```

```python
z = 1/10 * y**2
plt.scatter(x,z)
```

```python
plt.hist(z)
```

```python
z = 1/10 * y**3
plt.scatter(x,z)
plt.figure()
plt.hist(z)
```
